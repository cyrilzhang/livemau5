```python
import nn_all
import nn_arch
import numpy as np
import matplotlib.pyplot as plt
from plot_losses import plot_losses
#%matplotlib inline
```

```python
data = nn_all.load_data()
```

```python
nn_all.prepare(data, lambda x: np.mean(x, axis=0), directory="mean_flatten")
```

```python
nn_all.prepare(data, lambda x: np.max(x, axis=0) - np.min(x, axis=0), directory="range_flatten")
```

```python
nn_all.train_classify(num_epochs=1, net=nn_arch.onelayer_7x7, num_hidden_nodes=10, 
                  directory="mean_flatten", out_directory="mean_flatten/onelayer_7x7")
```

```python
nn_all.train_classify(num_epochs=1, net=nn_arch.onelayer_7x7, num_hidden_nodes=10, 
                  directory="range_flatten", out_directory="range_flatten/onelayer_7x7")
```

```python
nn_all.train_classify(num_epochs=100, net=nn_arch.twolayer_5x64, num_hidden_nodes=100,
                      directory="range_flatten", out_directory="range_flatten/twolayer_5x64_100fc",
                      restore=True, ckpt_directory="range_flatten/twolayer_5x64_100fc")
```

```python
nn_all.train_classify(num_epochs=100, net=nn_arch.threelayer_753x32, num_hidden_nodes=50,
                      directory="range_flatten", out_directory="range_flatten/threelayer_753x32_50fc",
                      restore=True, ckpt_directory="range_flatten/threelayer_753x32_50fc")
```

```python
nn_all.train_classify(num_epochs=100, net=nn_arch.fivelayer_3x24, num_hidden_nodes=50,
                      directory="range_flatten", out_directory="range_flatten/fivelayer_3x24_50fc",
                      restore=True, ckpt_directory="range_flatten/fivelayer_3x24_50fc")
```
